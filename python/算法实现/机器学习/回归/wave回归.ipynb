```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import mglearn

X,y = mglearn.datasets.make_wave(n_samples = 40)
plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel("Feature")
plt.ylabel("Target")
```

```python
mglearn.plots.plot_knn_regression(n_neighbors = 1)
```

```python
mglearn.plots.plot_knn_regression(n_neighbors = 3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor#k
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0)
reg = KNeighborsRegressor(n_neighbors = 3)
reg.fit(X_train,y_train)
reg.predict(X_test)
print("Test set predictions:\n {:.2f}".format(reg.score(X_test,y_test)))
```

```python
from sklearn.linear_model import LinearRegression
X,y = mglearn.datasets.make_wave(n_samples = 60)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42)
lr = LinearRegression()
lr.fit(X_train,y_train)
print(": {:.2f}".format((lr.coef_)[0]))
print(": {:.3f}".format(lr.intercept_))
print(": {:.2f}".format((lr.score(X_train,y_train))))
print(": {:.2f}".format((lr.score(X_test,y_test))))
```
